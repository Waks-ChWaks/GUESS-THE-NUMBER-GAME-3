```python
import random

# 🎮 Number Guessing Game
print("WOZZA! Guess The Number Ndio Mali Ya Leo!")
print("Siri Mbana Iko Between 1 and 20\n")

siri_mbana = random.randint(1, 20)
tries = 0
max_tries = 3

while tries < max_tries:
    guess = int(input("Nchapie Guess Yako: "))
    tries += 1

    if guess == siri_mbana:
        print("🎉 BAASS! Weh Ndio Unang'am Sasa!")
        break
    elif guess < siri_mbana:
        print("Iko chini tu sana. Uskatens. Raundi Ingine?")
    else:
        print("Iko juu tu sana. Uskatens. Raundi Ingine?.")

    if tries == max_tries:
        print(f"IZZA Kiongos. Betslip Imechomeka. Hio Time Yote Siri Mbana Ni {siri_mbana}.")
```

```python
import random
def guess_the_number():

    print("\n====== Guess The Number Game ======")
    print("\nI'm thinking of a number between 1 - 10.")
    print("\nTry to guess the number.")
    print("\nYou have 3 attempts!")

    secret_number = random.randint(1, 10)

    for attempt in range(1, 4):

        print()
        guess = int(input())

        if guess == secret_number:
            print("Congratulations!!!! You Guessed Right!!!")
            break
        elif guess > secret_number:
            print("Oops, too high!!! Guess a lower number. Attempts left =", 3 - attempt)
        else:
            print("Oops, too low!!! Guess a higher number. Attempts left =", 3 - attempt)

    print("\nGame Over!")
    print("The Secret Number Is:", secret_number)
    print("If You Didn't Find It, Try Again. May The Luck Be With You Next Time.")
guess_the_number()
```
